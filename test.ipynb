```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('IMG_6507.JPG')

plt.imshow(img)
```

```python
cv2.line(img, (0, 0), (4000, 4000), (255, 0, 0), 5)
cv2.rectangle(img, (0, 0), (4100, 4000), (0, 255, 0), 5)
cv2.circle(img, (1000, 1000), 50, (0, 255, 255), 5)

pts = np.array([[100, 50], [200, 300], [700, 200], [500, 100]], np.int32)
pts = pts.reshape((-1, 1, 2))
cv2.polylines(img, [pts], True, (255, 0, 255), 5)

cv2.putText(img, 'Hello World!', (3000, 2000), cv2.FONT_HERSHEY_SIMPLEX, 100, (0, 255, 255), 2, cv2.LINE_AA)

plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # convert to grayscale
edges = cv2.Canny(gray, 50, 150, apertureSize=3) # detect edges
lines = cv2.HoughLinesP(
                edges,
                1,
                np.pi/180,
                100,
                minLineLength=100,
                maxLineGap=100,
        ) # detect lines

for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(img, (x1, y1), (x2, y2), (0, 255, 0), 6)

plt.imshow(img)
```
